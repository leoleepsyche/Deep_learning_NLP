```python
import matplotlib.pylab as plt 
import numpy as np
def step_function(x):
    return np.array(x>0,dtype=np.int64)
x = np.arange(-5.0,5.0,0.1)
y = step_function(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```

### Sigmoid Function

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.array([-1.0,1.0,2.0])

y = sigmoid(x)
y
```

```python
t = np.array([1.0,2.0,3.0])
1.0 + t
```

```python
1.0 / t
```

```python
x = np.arange(-5.0,5.0,0.1)
y = sigmoid(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```

### ReLU Function (Rectified Linear Unit)

```python
def relu(x):
    return np.maximum(0,x)
```

```python
x = np.arange(-5.0,5.0,0.1)
y = relu(x)
plt.plot(x,y)
plt.ylim(-1.0,5.5)
plt.show()
```
